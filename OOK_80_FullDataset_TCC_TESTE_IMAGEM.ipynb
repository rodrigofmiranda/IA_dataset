```python
from PIL import Image
import matplotlib.pyplot as plt
import io

def save_figure(img):
  '''Função que salva a imagem e transforma em um array do np
  '''
  img_buf = io.BytesIO()
  img.savefig(img_buf, format='png')

  image = Image.open(img_buf).convert('L')
  image = image.resize((56, 56))
  numpy_image = np.asarray(image)
  img_buf.close()
  return numpy_image
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import io

df0 = pd.DataFrame(datas[0])
df30 = pd.DataFrame(datas[35])

numpy_img0 = []
numpy_img30 = []
#Cria imagem e coloca no array
for i in range(len(df0.columns)):
  fig0 = df0.iloc[:,i].plot(legend=False, cmap='gray')
  plt.axis('off')
  fig0 = fig0.get_figure()
  plt.close('all')
  fig30 = df30.iloc[:,i].plot(legend=False,cmap='gray')
  plt.axis('off')
  fig30 = fig30.get_figure()

  img0 = save_figure(fig0)
  img30 = save_figure(fig30)

  numpy_img0.append(img0)
  numpy_img30.append(img30)
  plt.close('all')
```

```python
np_img0 = np.array(numpy_img0)
np_img30 = np.array(numpy_img30)

X_train = np_img0[:4500,:,:]/255
X_train_noisy = np_img30[:4500,:,:]/255

X_test = np_img0[4500:-1,:,:]/255
X_test_noisy = np_img30[4500:-1,:,:]/255
```

```python
from tensorflow import keras
from keras import layers

'''conv_encoder = keras.models.Sequential([
    keras.layers.Reshape([28, 28, 1], input_shape=[28, 28]),
    keras.layers.Conv2D(16, kernel_size=3, padding="SAME", activation="selu"),
    keras.layers.MaxPool2D(pool_size=2),
    keras.layers.Conv2D(32, kernel_size=3, padding="SAME", activation="selu"),
    keras.layers.MaxPool2D(pool_size=2),
    keras.layers.Conv2D(64, kernel_size=3, padding="SAME", activation="selu"),
    keras.layers.MaxPool2D(pool_size=2)
])
conv_decoder = keras.models.Sequential([
    keras.layers.Conv2DTranspose(32, kernel_size=3, strides=2, padding="VALID", activation="selu",
                                 input_shape=[3, 3, 64]),
    keras.layers.Conv2DTranspose(16, kernel_size=3, strides=2, padding="SAME", activation="selu"),
    keras.layers.Conv2DTranspose(1, kernel_size=3, strides=2, padding="SAME", activation="sigmoid"),
    keras.layers.Reshape([28, 28])
])
conv_ae = keras.models.Sequential([conv_encoder, conv_decoder])

conv_ae.compile(loss="binary_crossentropy", optimizer=keras.optimizers.SGD(lr=1.0),
                metrics=['accuracy'])
history = conv_ae.fit(X_train_noisy, X_train, epochs=5,
                      validation_data=[X_test_noisy, X_test])

conv_encoder.summary()
conv_decoder.summary()
'''

model = keras.Sequential()
# encoder network
model.add(layers.Reshape([56, 56, 1], input_shape=[56, 56]))
model.add(layers.Conv2D(35, 3, activation= 'relu', padding='same', input_shape = (56, 56,1)))
model.add(layers.MaxPooling2D(2, padding= 'same'))
model.add(layers.Conv2D(25, 3, activation= 'relu', padding='same'))
model.add(layers.MaxPooling2D(2, padding= 'same'))
#decoder network
model.add(layers.Conv2DTranspose(25, 3, activation= 'relu', padding='same'))
model.add(layers.UpSampling2D(2))
model.add(layers.Conv2DTranspose(35, 3, activation= 'relu', padding='same'))
model.add(layers.UpSampling2D(2))
model.add(layers.Conv2DTranspose(1,3,activation='sigmoid', padding= 'same')) # output layer
model.add(layers.Reshape([56, 56]))

model.compile(optimizer= 'adam', loss = 'binary_crossentropy',metrics=['accuracy'])
model.fit(X_train_noisy, X_train,
                epochs=8,
                batch_size=128,
                validation_data=(X_test_noisy, X_test))
```
